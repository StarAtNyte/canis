```python
platforms = ['X (Twitter) Follower #', 'Facebook Follower #', 'Instagram Follower #',
             'Threads Follower #', 'YouTube Subscriber #', 'TikTok Subscriber #']

# Group by entity owner and calculate the total followers across all platforms
entity_owner_followers = df.groupby('Entity owner (English)')[platforms].sum()

# Sort by total followers and select the top 20
top_20_entity_owners = entity_owner_followers.sum(axis=1).nlargest(20).index

# Filter the dataframe to include only the top 20 entity owners
df_top_20 = df[df['Entity owner (English)'].isin(top_20_entity_owners)]

# Create a scatter plot for each platform
for platform in platforms:
    sns.scatterplot(x=platform, y='Entity owner (English)', data=df_top_20, label=platform)

# Customize the plot
plt.title("Top 20 Entity Owners: Social Media Influence")
plt.xlabel("Number of Followers/Subscribers")
plt.ylabel("Entity Owner")
plt.legend()

# Show the plot
plt.show()
```

```python
top_20_regions = df_top_20['Region of Focus'].value_counts().nlargest(20).index

df_top_20_filtered = df_top_20[df_top_20['Region of Focus'].isin(top_20_regions)]

sns.scatterplot(x='Region of Focus', y='Entity owner (English)',palette= 'colorblind', hue = 'Region of Focus',data=df_top_20_filtered)

plt.xlabel('Regions of Focus')
plt.ylabel('Entity Owners')
plt.title('Top 20 Entity Owners and their Regions of Focus')
plt.xticks(rotation=60, ha='right')

# Show the plot
plt.show()
```

```python
grouped_df = df.groupby(['Entity owner (English)', 'Parent entity (English)'])['Total Follower Counts'].sum()

# Create a new DataFrame with the summed follower counts
result_df = pd.DataFrame(grouped_df).reset_index()

result_df
```

```python
sorted_df = result_df.sort_values(by='Total Follower Counts', ascending=False)
top_40_entities = sorted_df.head(40)
top_40_entities
```

```python
top_40_entities = top_40_entities[top_40_entities['Parent entity (English)'] != 'unknown']
```

```python
df = top_40_entities[:10]
G = nx.DiGraph()

# Add nodes and edges to the graph
for index, row in df.iterrows():
    G.add_node(row['Entity owner (English)'], entity_type='Entity Owner')
    G.add_node(row['Parent entity (English)'], entity_type='Parent Entity')
    G.add_edge(row['Parent entity (English)'], row['Entity owner (English)'])

# Draw the graph
plt.figure(figsize=(12, 8))

# Separate nodes by entity type for coloring
entity_owner_nodes = [node for node, data in G.nodes(data=True) if data['entity_type'] == 'Entity Owner']
parent_entity_nodes = [node for node, data in G.nodes(data=True) if data['entity_type'] == 'Parent Entity']

# Increase the space between nodes by adjusting k
pos = nx.spring_layout(G, k=0.2)

nx.draw_networkx_nodes(G, pos, nodelist=entity_owner_nodes, node_color='lightblue', node_size=1000, label='Entity Owners')
nx.draw_networkx_nodes(G, pos, nodelist=parent_entity_nodes, node_color='lightcoral', node_size=1000, label='Parent Entities')
nx.draw_networkx_edges(G, pos, edge_color='blue', width=1.0)
nx.draw_networkx_labels(G, pos, labels=None, font_size=6, font_color='black', font_weight='bold')

# Add legend
plt.legend()

plt.title('Top 10 Entity Owners(According to Followers) and their Parent Entities Network Graph')
plt.show()
```

```python
grouped_df = df.groupby(['Parent entity (English)','Entity owner (English)'])['Total Follower Counts'].sum()

# Create a new DataFrame with the summed follower counts
result_df = pd.DataFrame(grouped_df).reset_index()

result_df.head()
```

```python
sorted_df = result_df.sort_values(by='Total Follower Counts', ascending=False)
sorted_df
```

```python

df_sorted = sorted_df.sort_values(by='Total Follower Counts', ascending=False)

# Get the top 10 parent entities
top_10_entities = df_sorted['Parent entity (English)'].head(10)

# Filter the dataframe for the top 10 entities
df_top_10 = df_sorted[df_sorted['Parent entity (English)'].isin(top_10_entities)]

# Pivot the dataframe to prepare for stacking
df_pivot = df_top_10.pivot(index='Parent entity (English)', columns='Entity owner (English)', values='Total Follower Counts')

# Sort columns based on the total values in descending order
df_pivot = df_pivot[df_pivot.sum(axis=0).sort_values(ascending=False).index]

# Plotting the stacked chart
ax = df_pivot.plot(kind='bar', stacked=True, figsize=(12, 8), colormap='RdYlBu', edgecolor='black', linewidth=0.5)

ax.set_ylabel('Total Follower Counts')
ax.set_title('Top 10 Parent Entities by Followers')

# Move the legend to the right
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5), title='Entity Owner')

# Show the plot
plt.show()
```
